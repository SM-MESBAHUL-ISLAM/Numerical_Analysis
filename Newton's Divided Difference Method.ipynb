```python
def linear_interpolation(x_list,y_list,xf):
    f1 = lambda b0,b1,xf,x0: b0 + b1*(xf - x0)
    
    for i in range(0,len(x_list)):
        if x_list[i] > xf:
            nearbyPoint = i
            break
    px0 = x_list[nearbyPoint-1]
    px1 = x_list[nearbyPoint]
    
    py0 = y_list[nearbyPoint-1]
    py1 = y_list[nearbyPoint]
    
    b0 = py0
    b1 = (py1 - py0)/(px1-px0)
    result = f1(b0,b1,xf,px0)
    return result

def quadratic_interpolation(x_list,y_list,xf):
    f2 = lambda b0,b1,b2,xf,x0,x1: b0 + b1*(xf-x0) + b2*(xf-x0)*(xf-x1)
    
    for i in range(0,len(x_list)):
        if x_list[i] > xf:
            nearbyPoint = i
            break
            
    px0 = x_list[nearbyPoint-2]
    px1 = x_list[nearbyPoint-1]
    px2 = x_list[nearbyPoint]
    py0 = y_list[nearbyPoint-2]
    py1 = y_list[nearbyPoint-1]
    py2 = y_list[nearbyPoint]
    
    b0 = py0
    b1 = (py1-py0)/(px1-px0)
    b2 = (((py2-py1)/(px2-px1))-((py1-py0)/(px1-px0)))/(px2-px0)
    
    result = f2(b0,b1,b2,xf,px0,px1)
    return result
    
def cubic_interpolation(x_list,y_list,xf):
    f3 = lambda b0,b1,b2,b3,xf,x0,x1,x2: b0 + b1*(xf-x0) + b2*(xf-x0)*(xf-x1) + b3*(xf-x0)*(xf-x1)*(xf-x2)
    
    for i in range(0,len(x_list)):
        if x_list[i] > xf:
            nearbyPoint = i
            break
    px0 = x_list[nearbyPoint-2]
    px1 = x_list[nearbyPoint-1]
    px2 = x_list[nearbyPoint]
    px3 = x_list[nearbyPoint+1]
    
    py0 = y_list[nearbyPoint-2]
    py1 = y_list[nearbyPoint-1]
    py2 = y_list[nearbyPoint]
    py3 = y_list[nearbyPoint+1]
    
    b0 = py0
    b1 = (py1-py0)/(px1-px0)
    b2 = (((py2-py1)/(px2-px1))-((py1-py0)/(px1-px0)))/(px2-px0)
    
    #finding b3:
    f_3_2 = (py3-py2)/(px3-px2)
    f_2_1 = (py2-py1)/(px2-px1)
    f_3_2_1 = (f_3_2 - f_2_1)/(px3-px1)
    f_2_1_0 = b2
    b3 = (f_3_2_1 - f_2_1_0)/(px3-px0)
    
    result = f3(b0,b1,b2,b3,xf,px0,px1,px2)
    return result

def comparison(x_list,y_list,xf):
    ans_of_linear = linear_interpolation(x_list,y_list,xf)
    ans_of_quadratic = quadratic_interpolation(x_list,y_list,xf)
    ans_of_cubic = cubic_interpolation(x_list,y_list,xf)
    
    # comparison table
    print("Order of Polynomial\t\t   :Linear \t\t Quadratic \t\t Cubic")
    e2 = (abs(ans_of_linear - ans_of_quadratic)/ans_of_linear)*100
    e3 = (abs(ans_of_quadratic - ans_of_cubic)/ans_of_quadratic)*100
    
    print("y(%f)\t\t\t   :%f\t\t %f \t\t %f"%(xf,ans_of_linear,ans_of_quadratic,ans_of_cubic))
    print("Absolute Relative approximate error:---------\t\t %.5f \t\t %.5f"%(e2,e3))

def newton_divided_difference_method():
    #this portion is for user input:
#     x_list = list(map(float,input("Enter the values of x: ").strip().split()))
#     y_list = list(map(float,input("Enter the values of y: ").strip().split()))
#     xf = float(input("Enter the calculation point xf: "))
    
    x_list = [0,10,15,20,22.5,30]
    y_list = [0,227.04,362.78,517.35,602.97,901.67]
    xf = 16
    print("The result from linear interpolation: ",linear_interpolation(x_list,y_list,xf))
    print("The result from quadratic interpolation: ",quadratic_interpolation(x_list,y_list,xf))
    print("The result from cubic interpolation: ",cubic_interpolation(x_list,y_list,xf))
    print("\nComparison table: ")
    comparison(x_list,y_list,xf)
    
newton_divided_difference_method()
```
