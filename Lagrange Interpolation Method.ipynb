```python
import sympy as sy

def linear_interpolation(x_list,y_list,xf):
    
    L = lambda x,xi,xj: (x-xj)/(xi-xj)
    
    for i in range(0,len(x_list)):
        if x_list[i] > xf:
            nearbyPoint = i
            break
    px0 = x_list[nearbyPoint]
    px1 = x_list[nearbyPoint-1]
    py0 = y_list[nearbyPoint]
    py1 = y_list[nearbyPoint-1]
    
    result = L(xf,px0,px1)*py0 + L(xf,px1,px0)*py1
    return result

def quadratic_interpolation(x_list,y_list,xf):
    
    L = lambda x,xi,xj,xk: ((x-xj)/(xi-xj))*((x-xk)/(xi-xk))
    
    for i in range(0,len(x_list)):
        if x_list[i] > xf:
            nearbyPoint = i
            break
    px0 = x_list[nearbyPoint-2]
    px1 = x_list[nearbyPoint-1]
    px2 = x_list[nearbyPoint]
    py0 = y_list[nearbyPoint-2]
    py1 = y_list[nearbyPoint-1]
    py2 = y_list[nearbyPoint]
    
    result = L(xf,px0,px1,px2)*py0 + L(xf,px1,px0,px2)*py1 + L(xf,px2,px0,px1)*py2
    return result

def cubic_interpolation(x_list,y_list,xf):
    
    L = lambda x,xi,xj,xk,xl: ((x-xj)/(xi-xj))*((x-xk)/(xi-xk))*((x-xl)/(xi-xl))
    
    for i in range(0,len(x_list)):
        if x_list[i]>xf:
            nearbyPoint = i
            break
    px0 = x_list[nearbyPoint-2]
    px1 = x_list[nearbyPoint-1]
    px2 = x_list[nearbyPoint]
    px3 = x_list[nearbyPoint+1]
    py0 = y_list[nearbyPoint-2]
    py1 = y_list[nearbyPoint-1]
    py2 = y_list[nearbyPoint]
    py3 = y_list[nearbyPoint+1]
    
    result = L(xf,px0,px1,px2,px3)*py0 + L(xf,px1,px0,px2,px3)*py1 + L(xf,px2,px0,px1,px3)*py2 + L(xf,px3,px0,px1,px2)*py3
    return result

def comparison(x_list,y_list,xf):
    ans_of_linear = linear_interpolation(x_list,y_list,xf)
    ans_of_quadratic = quadratic_interpolation(x_list,y_list,xf)
    ans_of_cubic = cubic_interpolation(x_list,y_list,xf)
    
    # comparison table
    print("Order of Polynomial\t\t   :Linear \t\t Quadratic \t\t Cubic")
    e2 = (abs(ans_of_linear - ans_of_quadratic)/ans_of_linear)*100
    e3 = (abs(ans_of_quadratic - ans_of_cubic)/ans_of_quadratic)*100
    
    print("y(%f)\t\t\t   :%f\t\t %f \t\t %f"%(xf,ans_of_linear,ans_of_quadratic,ans_of_cubic))
    print("Absolute Relative approximate error:---------\t\t %.5f \t\t %.5f"%(e2,e3))

def lagrange_interpolation_method():
    #this portion is for user input:
#     x_list = list(map(float,input("Enter the x values: ").strip().split()))
#     y_list = list(map(float,input("Enter the y values: ").strip().split()))
#     xf = float(input("Enter the calculation point xf: "))
    
    x_list = [0,10,15,20,22.5,30]
    y_list = [0,227.04,362.78,517.35,602.97,901.67]
    xf = 16
    print("The result from linear interpolation method is : ",linear_interpolation(x_list,y_list,xf))
    print("The result from quadratic interpolation method is : ",quadratic_interpolation(x_list,y_list,xf))
    print("The result from cubic interpolation method is : ",cubic_interpolation(x_list,y_list,xf))
    print("\nComparison table: ")
    comparison(x_list,y_list,xf)
    
lagrange_interpolation_method()
```
